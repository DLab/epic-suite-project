## Data Fit

```python
beta_values = [0.2,0.3,0.15,0.4]
beta_days = [50,65,80]
mu = 2.5
```

```python
cfg = 'SEIR.toml'
beta = cv19functions.piecewise(values = beta_values,limits=beta_days)

sims = CV19SIM(config=cfg,beta=beta,mu=mu,t_end=200)
sims.solve()

sim = sims.sims[0]
I_d_data = sim.I_d
t_data = sim.t
```
