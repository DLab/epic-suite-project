```python
# Working with pandas dataframe}
aux = []
for key in json.loads(response['results']['model2']).keys():
    aux.append(pd.DataFrame(json.loads(response['results']['model2'])[key]))
responsedf = pd.concat(aux)
```

```python
variable = 'I'
responsedf.groupby(by='node',axis=0)[variable].plot(legend=True,title=str(variable))
variable2 = 'S'
responsedf.groupby(by='node',axis=0)[variable2].plot(legend=True,title=str(variable + ' and ' + variable2),linestyle='dashed')
```
